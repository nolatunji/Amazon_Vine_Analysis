```python
# Read in customer reviews csv
df = pd.read_csv("vine_table.csv")
```

```python
df
```

```python
# Create a DataFrame where there are 20 or more total votes
vote_plus20_df = df[df.total_votes>20]
vote_plus20_df
```

```python
# Create a DataFrame where the percentage of 'helpful_votes' is equal to or greater than 50%

helpful_votes_df = vote_plus20_df[(vote_plus20_df.helpful_votes/df.total_votes)>=0.5]
helpful_votes_df
```

```python
# Filter data to create a DataFrame where there is a Vine review

vine_reviews_df = helpful_votes_df[helpful_votes_df.verified_purchase == 'Y']
vine_reviews_df
```

```python
# Filter data to create a DataFrame where there is not a Vine review

non_vine_reviews_df = helpful_votes_df[helpful_votes_df.verified_purchase == 'N']
non_vine_reviews_df
```

```python
# Determine the total 5-star Vine reviews

vine_reviews_df['star_rating'].value_counts()[5]
```

```python
# Determine the total 5-star non-Vine reviews

non_vine_reviews_df['star_rating'].value_counts()[5]
```

```python
# Determine the percentage of 5-star Vine reviews

percent_vine_reviews = (vine_reviews_df['star_rating'].value_counts()[5]/vine_reviews_df["review_id"].count())*100
percent_vine_reviews
```

```python
# Determine the percentage of 5-star non-Vine reviews

percent_nonvine_reviews = (non_vine_reviews_df['star_rating'].value_counts()[5]/non_vine_reviews_df["review_id"].count())*100
percent_nonvine_reviews
```

```python
# Determine the number of total customer reviews

total_reviews = helpful_votes_df["review_id"].count()
total_reviews
```

```python
# Determine the total number of 5-star reviews
five_stars = helpful_votes_df['star_rating'].value_counts()[5]
five_stars
```
